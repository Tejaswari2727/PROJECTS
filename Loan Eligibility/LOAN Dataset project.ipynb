```python
import matplotlib.pyplot as plt

# Calculate the frequency of each gender
gender_counts = loan_data['Gender'].value_counts()

# Here rename each bar according to 'Male' and 'Female' labels
gender_counts.index = ['Male', 'Female'] 

# Calculate the education frequency for each gender
education_counts = loan_data.groupby('Gender')['Education'].value_counts().unstack()

# Plotting the bar plot
education_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])

# Adding title and labels
plt.title('Education Frequency by Gender')
plt.xlabel('Gender')
plt.ylabel('Education Frequency')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the frequency of each marital status
marital_counts = loan_data['Married'].value_counts()

# Here rename each bar according to the marital status labels
marital_counts.index = ["Married","Unmarried"]  # Adjust according to your actual labels

# Calculate the education frequency for each marital status
education_counts = loan_data.groupby('Married')['Education'].value_counts().unstack()

# Plotting the bar plot
education_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'orange', 'pink'])

# Adding title and labels
plt.title('Education Frequency by Married')
plt.xlabel('Married')
plt.ylabel('Education Frequency')

# Displaying the plot
plt.show()
#more graduates are getting married
#non graduates are getting married very less in number
```

```python
loan_approval_percentages = loan_data.groupby('Self_emp')['Approved'].value_counts(normalize=True).unstack() * 100

ax = loan_approval_percentages.plot(kind='bar', stacked=True, color=['skyblue', 'green'], figsize=(8, 6))

plt.title('Percentage of Loan Approval by Self-Employment')
plt.xlabel('Self-Employed')
plt.ylabel('Percentage of Approved')
plt.xticks(rotation=0)  


for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.1f}%', (x + width / 2, y + height / 2), ha='center', va='center')

# Show plot
plt.show()
```
